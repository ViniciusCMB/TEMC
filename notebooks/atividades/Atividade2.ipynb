1. Crie um DataFrame que tenha 3 colunas: Produto |
Quantidade | Preço. Como valores, tenha duas linhas
com os seguintes valores na primeira linha: Chocolate |
200 | 3,00. Na segunda linha os valores: Banana | 80 |
0,50

```python
df = pd.DataFrame({
    'Produto': ['Chocolate', 'Banana'],
    'Quantidade': [200, 80],
    'Preço': [3.00, 0.50]
})

df
```

2. Crie uma Serie que contenha o total dos produtos:
Quantidade*Preço, ou seja, com os seguintes valores
600, 40.

```python
serie = pd.Series([df['Quantidade'][i]*df['Preço'][i] for i in range(len(df))])

serie
```
